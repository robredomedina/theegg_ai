```python
def count_characters(text):
    result = re.findall("[^\s]", text)
    print("There are {} characters".format(len(result)))
```

```python
def count_words(text):
    words = re.findall("\s?\w+[\s\W]", text)
    words_stripped = []
    for word in words:
        word = word.lstrip()
        while not word[-1].isalnum():
            word = word[:-1]
        words_stripped.append(word)
    print("There are {} words in the text".format(len(words_stripped)))
    return words_stripped
```

```python
def count_repetitions(words):
    words_count = {}
    for word in words:
        if not words_count.get(word):
            words_count[word] = 1
        else: 
            words_count[word] += 1
    return words_count
```

```python
def sort(words):
    for w in sorted(words, key=words.get, reverse=True):
        print("{0}: {1}".format(w, words[w]))
```
